```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic-Dataset.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df.Age.mean())
```

```python
df['Age'].isnull().sum()
```

```python
import scipy.stats as stats
```

```python
def diagnostic_plots(df, variable):
    # function to plot a histogram and a Q-Q plot
    # side by side, for a certain variable
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    df[variable].hist()

    plt.subplot(1, 2, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)

    plt.show()
    
```

```python
diagnostic_plots(df, 'Age')
```

```python
diagnostic_plots(df, 'Fare')
```

# Logarithmic Transformation

```python
df['New_fare']=np.log(df['Fare']+1)
diagnostic_plots(df, 'New_fare')
```

# Reciprocal transformation

```python
df['New_fare']=1/(df['Fare']+1)
diagnostic_plots(df, 'New_fare')
```

# Sqaure Root Tansformation

```python
df['New-Fare']=df['Fare']**(1/2)
diagnostic_plots(df, 'New_fare')
```

# Exponential Tanformation

```python
df['New-Fare']=df['Fare']**(1/5)
diagnostic_plots(df, 'New_fare')
```

```python
# THE BEST TRANFORMATION WAS Logarithmic Transformation WHICH GAVE THE BEST NORMAL DISTRIBUTION
```
