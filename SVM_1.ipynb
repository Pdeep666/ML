```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df['species'].value_counts()
```

```python
df0=df[df.species=='setosa']
df1=df[df.species=='versicolor']
df2=df[df.species=='virginica']
```

```python
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.scatter(df0['sepal_length'],df0['sepal_width'],color='g',marker='+')
plt.scatter(df1['sepal_length'],df1['sepal_width'],color='b',marker='_')
plt.scatter(df2['sepal_length'],df2['sepal_width'],color='r',marker='*')
```

```python
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.scatter(df0['petal_length'],df0['petal_width'],color='g',marker='+')
plt.scatter(df1['petal_length'],df1['petal_width'],color='b',marker='_')
plt.scatter(df2['petal_length'],df2['petal_width'],color='r',marker='*')
```

```python
x=df.drop(columns=['species'],axis=1)
y=df.species
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```
